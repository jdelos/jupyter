# Testing ploting using pandas df in notebooks

```python
import pandas as pd
import sys
```

Load the data and converte it to time series using the pandas framework

```python
cpu_df = pd.read_csv('cpu.csv', parse_dates=['st_time'],usecols=['st_time','t10s'])
mem_df = pd.read_csv('mem.csv', parse_dates=['st_time'],usecols=['st_time','pm_rel'])
```

```python
cpu_df['st_time'] = pd.to_datetime(cpu_df['st_time'])
mem_df['st_time'] = pd.to_datetime(mem_df['st_time'])
ts_cpu = cpu_df.set_index('st_time').resample('30S')['t10s']
ts_mem = mem_df.set_index('st_time').resample('30S')['pm_rel']
ts_cpu.interpolate()
ts_mem.interpolate()
```

Laod the interactive ploting for notebooks using bokeh. Load the enviroment for TimeSeries ploting

```python
from bokeh.io import push_notebook, output_notebook
from bokeh.charts import TimeSeries, show
output_notebook()
```

Plot the two time series

```python
data=dict(ts_cpu=ts_cpu['t10s'],
         time=ts_cpu['st_time'])

tscpu = TimeSeries(data,
    x='time', y=['ts_cpu'],
    legend=True)

```

```python
# read in some stock data from the Yahoo Finance API
AAPL = pd.read_csv(
    "http://ichart.yahoo.com/table.csv?s=AAPL&a=0&b=1&c=2000&d=0&e=1&f=2010",
    parse_dates=['Date'])
MSFT = pd.read_csv(
    "http://ichart.yahoo.com/table.csv?s=MSFT&a=0&b=1&c=2000&d=0&e=1&f=2010",
    parse_dates=['Date'])
IBM = pd.read_csv(
    "http://ichart.yahoo.com/table.csv?s=IBM&a=0&b=1&c=2000&d=0&e=1&f=2010",
    parse_dates=['Date'])

data = dict(
    AAPL=AAPL['Adj Close'],
    Date=AAPL['Date'],
    MSFT=MSFT['Adj Close'],
    IBM=IBM['Adj Close'],
)

tsline = TimeSeries(data,
    x='Date', y=['IBM', 'MSFT', 'AAPL'],
    color=['IBM', 'MSFT', 'AAPL'], dash=['IBM', 'MSFT', 'AAPL'],
    title="Timeseries", ylabel='Stock Prices', legend=True)

tspoint = TimeSeries(data,
    x='Date', y=['IBM', 'MSFT', 'AAPL'],
    color=['IBM', 'MSFT', 'AAPL'], dash=['IBM', 'MSFT', 'AAPL'],
    builder_type='point', title="Timeseries Points",
    ylabel='Stock Prices', legend=True)


t = show(tsline, notebook_handle=True)
```
